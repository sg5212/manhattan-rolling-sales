## Part 5: Describe the kind of buildings that were sold

### *Show a visualization that allows comparison of the number of the kinds of units sold: one family homes, office buildings, and condominiums*

```python
import matplotlib.pyplot as plt

# variables to hold the number of properties of each category
one_family_ct = 0
office_ct = 0
condos_ct = 0
other_ct = 0

# loop through the 'BUILDING CLASS CATEGORY' column and count occurrences 
# for each category asked for
for b in df['BUILDING CLASS AT PRESENT']: 
    if isinstance(b, float): 
        other_ct += 1
    else: 
        # count number of 'one family homes'  
        if b[0] == 'A': 
            one_family_ct += 1
        # count number of 'office buildings'  
        elif b[0] == 'O': 
            office_ct += 1
        # count number of 'condos' 
        elif b[0] == 'R': 
            condos_ct += 1
        # count number of 'other' 
        else: 
            other_ct += 1
    

# add up total number of properties
total_ct = one_family_ct + office_ct + condos_ct + other_ct

# calculate number of percentage for each category 
one_family_percent = (one_family_ct/total_ct) * 100     
office_percent = (office_ct/total_ct) * 100
condos_percent = (condos_ct/total_ct) * 100
other_percent = (other_ct/total_ct) * 100

# show a visualization (pie chart) that allows comparison of different kinds of units sold

# labels of the categories displayed in the pie chart 
labels = [r'One Family Homes (0.8683%)', r'Office Buildings (1.7600%)', r'Condominiums (52.2408%)', r'Other (45.1308%)']
# sizes of each slice of the pie chart 
sizes = [one_family_percent, commercial_percent, condos_percent, other_percent]
# highlight (by "exploding") the largest slice
explode = (0, 0, 0.1, 0)
# assemble the pie
patches, texts = plt.pie(sizes, explode=explode, startangle=90)
# create a legend to match the pie slices to corresponding category 
# (instead of labels because labels were overlapping due to small slices)
plt.legend(patches, labels, loc='best')
# equal aspect ratio to ensure that pie is drawn as a circle
plt.axis('equal')
# adjust padding so pie chart nicely fits in the figure 
plt.tight_layout()
# display pie chart
plt.show()
```

## Part 6: Calculate summary statistics for the prices of properties sold for all of Manhattan and for a couple of select neighborhoods

### *Use any method to calculate mean, median, percentiles (25 and 75), max, and min*

```python
# clean data: convert column 'SALE PRICE' from object to float64 
df[' SALE PRICE '].value_counts()
def normalize_saleprice(sale_price): 
    # take out any non-numeric characters
    if isinstance(sale_price, float): 
        return sale_price
    if ',' in sale_price: 
        # split by commas and combine 
        arr_price = sale_price.split(',')
        str_price = ''
        for s in arr_price: 
            str_price += s
        
        flt_price = float(str_price)
        return flt_price; 
   
df[' SALE PRICE '] = df[' SALE PRICE '].map(normalize_saleprice)

# percentile25 = 0
# percentile25 = 0
# max_var = 0
# min_var = 0
total_price = 0

for price in df[' SALE PRICE ']: 
    total_price += price

print('Mean: ', df[' SALE PRICE '].mean())
print('Median: ', df[' SALE PRICE '].median())
print('25th percentile: ', df[' SALE PRICE '].quantile(0.25))
print('75th percentile: ', df[' SALE PRICE '].quantile(0.75))
print('Max: ', df[' SALE PRICE '].max())
print('Min: ', df[' SALE PRICE '].min())
```

### *Pick two neighborhoods: calculate summary statistics for each neighborhood: use any method to calculate mean, median, percentiles (25 and 75), max, and min*

```python
# Neighborhoods chosen: Flatiron & Chelsea
# df['NEIGHBORHOOD'].value_counts()

# summary statistics for Flatiron
flatiron_price = 0
flatiron_total = 0
flatiron_arr = []

# array of indices of rows that have the value of 'FLATIRON'
flatiron_index = df[df['NEIGHBORHOOD'] == 'FLATIRON'].index
# loop through these indices
for index in flatiron_index: 
    # get the corresponding 'SALE PRICE'
    flatiron_price = df.iloc[index][' SALE PRICE ']
    # add to a total variable to get the total price of all 'FLATIRON' properties     
    flatiron_total += flatiron_price
    # add to array to hold all the prices of 'FLATIRON' properties
    flatiron_arr.append(flatiron_price)
    
# display summary statistics of 'FLATIRON' using numpy functions
print('Displaying summary statistics for Flatiron')
print('Mean (Flatiron): ', np.mean(flatiron_arr))
print('Median (Flatiron): ', np.median(flatiron_arr))
print('25th Percentile (Flatiron): ', np.percentile(flatiron_arr, 25))
print('75th Percentile (Flatiron): ', np.percentile(flatiron_arr, 75))
print('Max (Flatiron): ', np.max(flatiron_arr))
print('Min (Flatiron): ', np.min(flatiron_arr), '\n\n')

# summary statistics for Chelsea
chelsea_price = 0
chelsea_total = 0
chelsea_arr = []

# array of indices of rows that have the value of 'CHELSEA'
chelsea_index = df[df['NEIGHBORHOOD'] == 'CHELSEA'].index
# loop through these indices
for index in chelsea_index: 
    # get the corresponding 'SALE PRICE'
    chelsea_price = df.iloc[index][' SALE PRICE ']
    # add to a total variable to get the total price of all 'CHELSEA' properties   
    chelsea_total += chelsea_price
    # add to array to hold all the prices of 'CHELSEA' properties
    chelsea_arr.append(chelsea_price)
    
# display summary statistics of 'CHELSEA' using numpy functions
print('Displaying summary statistics for Chelsea')
print('Mean (Chelsea): ', np.mean(chelsea_arr))
print('Median (Chelsea): ', np.median(chelsea_arr))
print('25th Percentile (Chelsea): ', np.percentile(chelsea_arr, 25))
print('75th Percentile (Chelsea): ', np.percentile(chelsea_arr, 75))
print('Max (Chelsea): ', np.max(chelsea_arr))
print('Min (Chelsea): ', np.min(chelsea_arr))

    
```

Comparing results: 

When looking at the summary statistics for the two neighborhoods I chose, Flatiron and Chelsea, I can conclude that Flatiron is a more expensive neighborhood as all the measures for Flatiron were higher than those of Chelsea. Although the mean of both neighbordhoods varied greatly, the medians were closer in value, indicating that Chelsea has a more symmetrical distribution than Flatiron. The 25th and 75th percentiles for both nieghborhoods are also close in value, however, the max of Chelsea is much greater than that of Flatiron.

## Part 7: Bin the prices of properties sold

### *Bin the prices and create a visualization that compares the bins*

```python
# create 5 bins to fit the prices of 'CHELSEA' and 'FLATIRON' prices

# combine both 'CHELSEA' and FLATIRON' prices arrays 
total_arr = flatiron_arr + chelsea_arr

# creating bins according to min, percentiles of total_arr, and max
total_min = min(total_arr)-1
total_percentile20 = np.percentile(total_arr, 20)
total_percentile40 = np.percentile(total_arr, 40)
total_percentile60 = np.percentile(total_arr, 60)
total_percentile80 = np.percentile(total_arr, 80)
total_max = max(total_arr)# array to hold the bin values
bins = [total_min, total_percentile20, total_percentile40, total_percentile60, total_percentile80, total_max]
# array to hold the bin labels 
labels = ['BIN 1 [0-96258.0)', 'BIN 2 [96258.0-840000.0)', 'BIN 3 [840000.0-1650000.0)', 'BIN 4 [1650000.0-3200000.0)', 'BIN 5 [3200000.0-880000000.0)']
# use cut to bin the prices into bins 
inbins = pd.cut(total_arr, bins, labels=labels)
# count values that fall into each bin
bins_sort = pd.value_counts(inbins)
print(bins_sort.sort_index())


# create a visualization that compares the bins 
# calculate percentages of each bin
bin1_percent = 193/len(total_arr)
bin2_percent = 192/len(total_arr)
bin3_percent = 192/len(total_arr)
bin4_percent = 190/len(total_arr)
bin5_percent = 189/len(total_arr)

# labels of the categories displayed in the pie chart 
labels = [r'BIN 1 [0-96258.0)', r'BIN 2 [96258.0-840000.0)', r'BIN 3 [840000.0-1650000.0)', r'BIN 4 [1650000.0-3200000.0)', r'BIN 5 [3200000.0-880000000.0)']
# sizes of each slice of the pie chart 
sizes = [bin1_percent, bin2_percent, bin3_percent, bin4_percent, bin5_percent]
# highlight (by "exploding") the largest slice
explode = (0.1, 0, 0, 0, 0)
# assemble the pie
patches, texts = plt.pie(sizes, explode=explode, startangle=90)
# create a legend to match the pie slices to corresponding category 
# (instead of labels because labels were overlapping due to small slices)
plt.legend(patches, labels, loc='best')
# equal aspect ratio to ensure that pie is drawn as a circle
plt.axis('equal')
# adjust padding so pie chart nicely fits in the figure 
plt.tight_layout()
# display pie chart
plt.show()
```

**What is unusual about the dataset's prices?**
Bin 2 and Bin 5 has the greatest range among all bins; however, the number of prices that fall in these bins are very similar to the number of prices in other bins. 


**How might this affect the earlier calculations?**
This would affect the means of the previous calculations since this indicates that there are outliers present in the data set.

## Part 8: Create a visualization the shows the relationship (if any) between the price sold and the number of square footage of all the areas of a property within a building or a structure

```python
# df['GROSS SQUARE FEET'].value_counts()
def normalize_gross_square_feet(gross_square_feet): 
    # take out any non-numeric characters
    if np.isnan(gross_square_feet): 
        return 0.0
    if isinstance(gross_square_feet, float): 
        return gross_square_feet
    if ',' in gross_square_feet: 
        # split by commas and combine 
        arr_feet = gross_square_feet.split(',')
        str_feet = ''
        for st in arr_feet: 
            str_feet += st
        
        flt_feet = float(str_feet)
        return flt_feet; 
   
df['GROSS SQUARE FEET'] = df['GROSS SQUARE FEET'].map(normalize_gross_square_feet)

# visualization the shows the relationship (if any) between the price sold and the number 
# of square footage of all the areas of a property within a building or a structure
x_price_sold = df[' SALE PRICE ']
y_gross_square_feet = df['GROSS SQUARE FEET']
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x_price_sold, y_gross_square_feet, color='b')
ax.set_xlabel('Price Sold')
ax.set_ylabel('Gross Square Footage')
ax.set_title('Price Sold vs. Gross Square Footage')
plt.show()
```
